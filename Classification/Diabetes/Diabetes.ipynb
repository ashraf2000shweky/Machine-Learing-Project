```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier # نموذج الجار الأقرب
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier ,GradientBoostingClassifier

from sklearn.pipeline import Pipeline # تطبيق خوازميات النماذج ضمن انابيب أي تطبيق متتالي للنماذج 

```

```python
data = pd.read_csv('diabetes.csv')
data
```

```python
data.isnull().sum()
```

```python
data.describe(include='all')
```

```python
data_copy = data.copy(deep=True)
```

```python
data.columns
```

```python
data_copy[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']] = data_copy[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']].replace(0,np.nan)
```

```python
data_copy.isnull().sum()
```

```python
data['Glucose'] = data['Glucose'].replace(0,data['Glucose'].median())
data['BloodPressure'] = data['BloodPressure'].replace(0,data['Glucose'].median())
data['SkinThickness'] = data['SkinThickness'].replace(0,data['Glucose'].median())
data['BMI'] = data['BMI'].replace(0,data['Glucose'].median())
data['Insulin'] = data['Insulin'].replace(0,data['Glucose'].median())

```

```python
X = data.drop('Outcome',axis = 1)
y = data['Outcome']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
# تطبيق متتالي خوازمية التحجيم ثم النموذج المحدد
pipeline_lr = Pipeline([('scalar1',StandardScaler()),('lr_classifier',LogisticRegression())])
pipeline_knn = Pipeline([('scalar2',StandardScaler()),('Knn_classifier',KNeighborsClassifier())])
pipeline_svc = Pipeline([('scalar3',StandardScaler()),('svc_classifier',SVC())])
pipeline_dt = Pipeline([('dt_classifier',DecisionTreeClassifier())])
pipeline_rf = Pipeline([('rf_classifier',RandomForestClassifier())])
pipeline_gbc = Pipeline([('gbc_classifier',GradientBoostingClassifier())])

```

```python
pipelines = [pipeline_lr ,pipeline_knn ,pipeline_svc,pipeline_dt, pipeline_rf, pipeline_gbc]
```

```python
pipelines
```

```python
# تركيب كل النماذج 
for pipe in pipelines:
    pipe.fit(X_train,y_train)
    
```

```python
pipe_dict = {0:'LR',
             1:'KNN',
             2:'SVC',
             3:'DT',
             4:'RF',
             5:'GBC'}
```

```python
# عرض تقييم النموذج 
for i,model in enumerate(pipelines):
    print("{} Test Accuracy: {}".format(pipe_dict[i],model.score(X_test,y_test)*100))
```
